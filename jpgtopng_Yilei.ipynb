```python
from PIL import Image
import os

def convert_and_overwrite(input_folder):
    # 递归遍历输入文件夹及其所有子文件夹中的文件
    for root, dirs, files in os.walk(input_folder):
        for filename in files:
            if filename.endswith(('.jpg', '.JPG')):
                # 构造输入和输出文件的完整路径
                input_path = os.path.join(root, filename)
                temporary_output_path = os.path.splitext(input_path)[0] + '.png'
                
                # 打开JPG图片并保存为PNG格式
                with Image.open(input_path) as img:
                    img.save(temporary_output_path, 'PNG')

                # 删除原始的JPG文件
                os.remove(input_path)

                # 将PNG文件重命名为原始JPG的文件名，但保留PNG扩展名
                new_png_filename = os.path.splitext(input_path)[0] + '.png'
                os.rename(temporary_output_path, new_png_filename)

                print(f"Converted {input_path} to {new_png_filename}")

# 指定输入文件夹
input_folder = r"D:\\GitHub\\Monovit\\kitti_data\\"

# 调用函数进行转换
convert_and_overwrite(input_folder)
```
