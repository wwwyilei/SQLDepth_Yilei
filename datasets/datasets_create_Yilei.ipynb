```python
import os

# 指定文件夹路径
folder_path = r"D:\Datasets\training"

# 获取文件夹中的所有文件名，并按照文件名排序
files = sorted(os.listdir(folder_path), key=lambda x: int(os.path.splitext(x)[0]))

# 遍历文件列表，对每个文件进行重命名
for i, file in enumerate(files):
    # 构造新文件名
    new_name = os.path.join(folder_path, f"{i}.png")
    # 旧文件完整路径
    old_name = os.path.join(folder_path, file)
    # 重命名文件
    os.rename(old_name, new_name)
    print(f"Renamed {old_name} to {new_name}")
```

```python
import os
import random

# 文件夹路径
folder_path = r"D:\Datasets\selfvideo\training"
# 训练集和验证集比例
train_ratio = 0.8
val_ratio = 0.2

# 获取文件夹中的所有文件名
files = os.listdir(folder_path)
# 将文件名按照原始顺序排序
files.sort()

# 计算训练集和验证集的数量
num_train = int(len(files) * train_ratio)
num_val = len(files) - num_train

# 随机选择训练集和验证集的文件名
train_files = random.sample(files, num_train)
val_files = [file for file in files if file not in train_files]

# 写入训练集txt文件
with open(r"D:\GitHub\SfMNeXt-Impl\splits\selfvideo\train_files.txt", "w") as f:
    for file in train_files:
        file_name = os.path.splitext(file)[0]  # 去除文件后缀
        f.write(f"selfvideo\\training {file_name} l\n")

# 写入验证集txt文件
with open(r"D:\GitHub\SfMNeXt-Impl\splits\selfvideo\val_files.txt", "w") as f:
    for file in val_files:
        file_name = os.path.splitext(file)[0]  # 去除文件后缀
        f.write(f"selfvideo\\training {file_name} l\n")
```
